```python
class tree:
    def __init__(self,data):
        self.right=None
        self.left=None
        self.main_data=data

    def insert(self,new_data):
        if self.main_data:
            if new_data<self.main_data:
                if self.left==None:
                    self.left=tree(new_data)
                else:
                    self.left.insert(new_data)
            elif new_data>self.main_data:
                if self.right==None:
                    self.right=tree(new_data)
                else:
                    self.right.insert(new_data)

        else:
            self.main_data=new_data

    def show(self):
        if self.left:
            self.left.show()
        print(self.main_data)
        if self.right:
            self.right.show()
```

```python
obj=tree(23)
obj.insert(11)
obj.insert(65)
obj.insert(3)
obj.insert(21)
obj.insert(12)
obj.insert(98)
obj.insert(1)
obj.insert(-21)
obj.insert(25)
obj.insert(-65)
obj.insert(-11)
obj.insert(91)
obj.insert(5)
```

```python
class stack:
    def __init__(self):
        self.lis=[]

    def insert(self,new_data):
        self.lis.append(new_data)

    def show(self):
        print(self.lis)

    def last_data(self):
        print(self.lis[-1])

    def take(self):
        return self.lis.pop()

    def rev(self):
        return self.lis[::-1]

```
